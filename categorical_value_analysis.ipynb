<font color='red'>Assignment (hard):</font> Many rows of **product_type**, **maker**, and **ingredient** contain multiple categorical values. For example, "P,R" in **product_type** actually indicates this product belongs to two different product types "P" and "R". Can you obtain the value counts for single **product_type**, **maker**, and **ingredient**?

```python
def mul_cat_val_cnt(input_ser):
    my_count = {}
    for i in range(len(input_ser)):
        if type(input_ser[i]) == str:
            my_str = input_ser[i].split(",")
            for j in range(len(my_str)):
                if my_str[j] not in my_count:
                    my_count[my_str[j]] = 0
                my_count[my_str[j]] += 1
    return my_count
        
```

```python
val_cnt1 = mul_cat_val_cnt(df.product_type)
val_cnt_product_type = pd.Series(val_cnt1)
val_cnt_product_type
```

```python
val_cnt1 = mul_cat_val_cnt(df.maker)
val_cnt_maker = pd.Series(val_cnt1)
val_cnt_maker
```

```python
val_cnt1 = mul_cat_val_cnt(df.ingredient)
val_cnt_ingredient = pd.Series(val_cnt1)
val_cnt_ingredient
```
